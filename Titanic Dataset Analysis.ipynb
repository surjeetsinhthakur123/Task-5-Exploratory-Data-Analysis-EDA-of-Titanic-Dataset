```python
train_df.hist(figsize=(15, 10), edgecolor='black')
plt.suptitle('Histograms of Numerical Features', fontsize=20)
plt.show()
```

```python
features = ['Age', 'Fare']
for feature in features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=train_df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
features = ['Age', 'Fare']
for feature in features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=test_df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
categorical_features = ['Sex', 'Embarked', 'Pclass', 'Survived']

for feature in categorical_features:
    print(f"\nValue counts for {feature}:\n{train_df[feature].value_counts()}")

    sns.countplot(x=feature, data=train_df, palette='pastel')
    plt.title(f'Countplot of {feature}')
    plt.show()
```

```python
categorical_features = ['Sex', 'Embarked', 'Pclass']

for feature in categorical_features:
    print(f"\nValue counts for {feature}:\n{test_df[feature].value_counts()}")

    sns.countplot(x=feature, data=test_df, palette='pastel')
    plt.title(f'Countplot of {feature}')
    plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=train_df, palette='coolwarm')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=train_df, palette='coolwarm')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Survived', data=train_df)
plt.title('Age vs Survival')
plt.show()
```

```python
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
corr = train_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
test_df['Age'].fillna(test_df['Age'].median(), inplace=True)
corr = test_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(train_df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='husl')
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()
```
